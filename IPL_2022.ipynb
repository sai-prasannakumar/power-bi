```python
import pandas as pd
import numpy as np
import datetime
import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_html("https://stats.espncricinfo.com/ci/engine/records/team/match_results.html?id=14452;type=tournament")
```

```python
res = df[0]
```

```python
res
```

```python
res['Margin'] = res['Margin'].apply(str)
res['Win_Method'] = np.where(res['Margin'].str.endswith('ns'),'chasing','defending')
```

```python
results = res[['Team 1', 'Team 2', 'Winner','Win_Method', 'Margin', 'Ground']]
```
